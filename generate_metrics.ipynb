```python
## ------------------ Imports ----------------------------------

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

## ------------------ Calculate Metrics ------------------------

def find_mean_price():
    #Returns the mean price of all the items
    data = pd.read_csv('things.csv', encoding='latin1')
    total_prices = 0
    price_count = 0
    mean_price = 0
    
    for i in data['PRICE']:
        if(i == 0):
            pass
        else:
            total_prices += i
            price_count += 1
            
    try:
        mean_price = (total_prices / price_count)
    except ZeroDivisionError:
        return 0
        
    return mean_price

def frame_data(searchword):
    #Will print out all mean values of price, rating, and feedback
    data = pd.read_csv('things.csv', encoding='latin1')
    
    mean_rating = 0
    mean_feedback = 0
    mean_price = find_mean_price()
    
    for x in data['RATING']:
        mean_rating += x
    for j in data['FEEDBACK']:
        mean_feedback += j
    
    mean_rating = mean_rating / len(data['RATING'])
    mean_feedback = mean_feedback / len(data['FEEDBACK'])
    print("The Mean Price is ${}".format(round(mean_price,2)))
    print("The Mean Amount of Ratings is {}".format(int(mean_rating)))
    print("The Mean Positive Feedback Percentage is {}%".format(round(mean_feedback,2)))
    
def create_histogram(category):
    #Will generate and show a histogram for the prices of the items
    data = pd.read_csv('things.csv', encoding='latin1')
    prices = data[category]
    plt.hist(prices.values, bins=25, edgecolor="k")
    plt.title(category)
    plt.show()
    
def top_five(best_prices, category):
    #Will find the top five ranked items in the given category
    data = pd.read_csv('things.csv', encoding='latin1')
    best = [[0,0]] * 5
    grouped_category = data[category]
    
    for i in best_prices:
        for x in range(len(best)):
            if(grouped_category[i] > best[x][1]):
                best[x][0] = i
                best[x][1] = grouped_category[i]
                break
                
    return best
    
def find_best_item():
    data = pd.read_csv('things.csv', encoding='latin1')
    mean_price = find_mean_price()
    best_prices = []
    prices = data['PRICE']
    best_items = {}
    
    #Go through every item and find items with price around the mean
    for i in range(len(prices)):
        if(abs(prices[i] - mean_price) <= 1):
            best_prices.append(i)
        
    #Build the dictionary of every item with the best price
    #Each items score will increase if their ranking in all other 
    #categories is in the top 5.
    for x in best_prices:
        best_items[x] = 0
    
    solds = data['SOLD']
    ratings = data['RATING']
    feedbacks = data['FEEDBACK']
    
    best_sold = top_five(best_prices, 'SOLD')
    best_rating = top_five(best_prices, 'RATING')
    best_feedback = top_five(best_prices, 'FEEDBACK')
                
    for i in range(len(best_sold)):
        try:
            best_items[best_sold[i][0]] = best_items[best_sold[i][0]] + 1
            best_items[best_rating[i][0]] = best_items[best_rating[i][0]] + 1
            best_items[best_feedback[i][0]] = best_items[best_feedback[i][0]] + 1
        except KeyError:
            pass
        
    top_item = [0,0]
    for i in best_prices:
        if(best_items[i] > top_item[1]):
            top_item = [i, best_items[i]]
    
    #This will be the index for the items url in "products" defined in main
    return top_item[0]
 
```
